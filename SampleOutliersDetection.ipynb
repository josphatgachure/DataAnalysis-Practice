```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
N = 10 ** 4
np.random.seed(42)
gaussian_data = np.random.normal(size=N)
skewed_data = np.random.lognormal(size=N)
sns.distplot(gaussian_data)
plt.title('Normalized_Data')
```
