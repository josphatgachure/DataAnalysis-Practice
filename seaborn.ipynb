```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
penguins = sns.load_dataset("penguins")
tips
```

```python
sns.set_theme()
sns.scatterplot(data=tips, x="total_bill",y="tip", hue="smoker")
```

```python
sns.scatterplot(data=tips, x="total_bill",y="tip", hue="day")
```

```python
sns.scatterplot(data=tips, x="total_bill",y="tip", hue="sex", style="time")
```

line plots

```python
flights=sns.load_dataset("flights")
flights
```

```python
sns.lineplot(data=flights,x ="year",y="passengers",estimator="sum" )
#the default behaviour is to take the mean and plot but the estimator property talls it what to plot,eg max,sum etc
flights.groupby("year")["passengers"].mean().plot() #= the first line without estimator
```

```python
sns.lineplot(data=flights,x ="year",y="passengers",hue="month" )
```

```python
trips = sns.load_dataset("taxis",parse_dates=["pickup","dropoff"])
trips["hour"]= trips["pickup"].dt.hour

sns.lineplot(data=trips,x="hour",y="total",hue="payment")
```

```python
sns.lineplot(data=trips,x="hour",y="total",hue="payment",style="color",ci=None)
```

```python
sns.relplot(data=tips,x="total_bill",y="tip", kind="scatter")
```

```python
sns.relplot(data=tips,x="total_bill",y="tip", kind="scatter" ,col="sex",hue="smoker")
```

```python
sns.relplot(data=tips,x="total_bill",y="tip", kind="scatter" ,col="sex",hue="smoker",row="time")
```
