```python
plants =['Spinach','Turnip','Rhubarb','Brocolli','Kales']
died = [10,25,5,30,21]
germinated = [74,88,56,69,59]
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,germinated,color="green")
plt.title("Germination Rate")
plt.ylabel("Number of Seedlings" ,labelpad=8)
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,germinated,color="olive", width=0.5)
plt.title("Germination Rate")
plt.ylabel("Number of Seedlings" ,labelpad=8)
```

```python
#multiple
plt.figure(figsize=(6,6))
plt.bar(plants,germinated,color="olive", width=0.5)
plt.title("Germination Rate")
plt.ylabel("Number of Seedlings" ,labelpad=8)
```

```python
#combinined 
plt.figure(figsize=(6,6))
plt.bar(plants,germinated,color="olive",width=0.5, align="edge")
plt.bar(plants,died,color="orchid",width=0.4)
plt.title("Combined Rates")
plt.ylabel("Number of Seedlings" ,labelpad=8)
```

```python
# A Stacked ber
plt.figure(figsize=(6,6))
plt.bar(plants,germinated,color="olive", label="Germinated")
plt.bar(plants,died,color="orchid",bottom=germinated,label="died")
plt.title("Combined Rates")
plt.ylabel("Number of Seedlings" ,labelpad=8)
plt.legend()
```

```python

plt.figure(figsize=(6,6))
plt.barh(plants,germinated,color="olive")
plt.title("Germination Rate")
plt.ylabel("Number of Seedlings" ,labelpad=8)
```

```python
import numpy as np
nums = np.arange(5)
plt.figure(figsize=(10,4))
plt.suptitle("OUR SUBPLOT")
plt.tight_layout()
plt.subplot(1,3,1)
plt.plot(nums, nums)
plt.title("x")
plt.subplot(1,3,2)
plt.plot(nums, nums**2,color="orchid")
plt.title("x squared")
plt.subplot(1,3,3)
plt.plot(nums, nums**3,color="olive")
plt.title("x cubed")
```

```python
import pandas as pd
titanic = pd.read_csv("train.csv")

first_class = titanic[titanic["Pclass"]==1]["Age"]
second_class = titanic[titanic["Pclass"]==2]["Age"]
third_class = titanic[titanic["Pclass"]==3]["Age"]

plt.figure(figsize=(10,4))
ax = plt.subplot(1,3,1)
plt.hist(first_class,label="1stclass",bins=20,color="brown")
plt.legend()
plt.title("1stclass")

plt.subplot(1,3,2,sharey=ax)
plt.hist(second_class,label="2ndclass",bins=20,color="olive")
plt.legend()
plt.title("2ndclass")

plt.subplot(1,3,3,sharey=ax)
plt.hist(third_class,label="3rdclass",bins=20,color="orchid")
plt.legend()
plt.title("3rdclass")
```
