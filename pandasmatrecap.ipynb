```python
plt.figure(figsize=(10,7))
xmas = chart[chart["date"].between("2016-12-25","2021-01-01")].set_index("date")
Bird = xmas[xmas["song"]=="All I Want For Christmas Is You"]
Bird["rank"].plot(label="All I want")
Rock = xmas[(xmas["song"]=="Rockin' Around The Christmas Tree")&  (xmas["artist"]=="Brenda Lee")]
Rock["rank"].plot(label = "Rocking")

Jingle = xmas[xmas["song"]=="Jingle Bell Rock"]
Jingle["rank"].plot(label="Jingle Bell")
plt.legend()
plt.title("Christmas Song")
plt.ylabel("chart rank")
plt.xticks(["2016-12-25","2017-12-25","2018-12-25","2019-12-25","2020-12-25"],
          labels=["Xmas 2016","Xmas 2017","Xmas 2018","Xmas 2019","Xmas 2020"])
plt.gca().invert_yaxis()

```

```python

jl= chart[chart["artist"]=="John Lennon"]["song"].value_counts().head(8)
Wings = chart[chart["artist"]=="Wings"]["song"].value_counts().head(8)
gh = chart[chart["artist"]=="George Harrison"]["song"].value_counts().head(8)
rs = chart[chart["artist"]=="Ringo Starr"]["song"].value_counts().head(8)

fig, axs=plt.subplots(2,2,figsize=(14,8))
fig.suptitle("Beatles Solos Songs(By week)", fontsize=20)

jl.plot(kind="barh", ax=axs[0][0])
axs[0][0].set_title("John Lenon",color="purple")

Wings.plot(kind="barh", ax=axs[0][1],color="olive")

axs[0][1].set_title("Wing",color="Red")
plt.tight_layout()#to avoid overlap


gh.plot(kind="barh", ax=axs[1][0],color="orchid")

axs[1][0].set_title("George",color="brown")

rs.plot(kind="barh", ax=axs[1][1],color="green")

axs[1][1].set_title("Starr",color="Brown")
plt.savefig("Beetles")
```
